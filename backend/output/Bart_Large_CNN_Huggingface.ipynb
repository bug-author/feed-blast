```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")
```

```python
texts=[]
```

```python
import json

with open('news.json', 'r', encoding='utf8') as in_file:
  data = json.load(in_file)

for news_list in data.values():
  for news_article in news_list:
    texts.append(news_article['details'])
```

```python
len(texts)
```

```python
summaries=[]
```

```python
def summarize(text, max_length=512):
  input_ids = tokenizer.encode(text[:512], return_tensors="pt", add_special_tokens=True)

  generated_ids = model.generate(input_ids=input_ids, num_beams=2, max_length=max_length,  repetition_penalty=2.5, length_penalty=1.0, early_stopping=True)

  preds = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=True) for g in generated_ids]

  return preds[0]
```

```python
i=1
for text in texts:
  tokens = tokenizer(text, truncation=True, padding='longest', return_tensors='pt')

  summary = summarize(text)

  summaries.append(summary)
  print(i)
  i+=1
```
